```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def readDense(filename):
    lines = open(filename,"r").readlines()
    for i,line in enumerate(lines):
        if line[0] != "%":
            break
    m,n = lines[i].split(' ')
    m,n = int(m),int(n)
    tab = np.zeros((n,m), dtype = np.uint32)
    for j in range(m*n):
        tab[j//m, j%m] = int(lines[j+i+1])
    return tab
```

```python
tab0 = readDense("recordings/Av0it1.mtx")
tab1 = readDense("recordings/Av1it1.mtx")
```

```python
plt.imshow(tab0)
plt.figure()
plt.imshow(tab1)
```

```python
tab = readDense("recordings/Avit1.mtx")
```

```python
np.linalg.norm(tab[:,:240]-tab0)
```
